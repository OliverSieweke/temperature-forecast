```python
from geopy.geocoders import Nominatim
import folium
```

```python
germany = folium.Map(location=[52.49, 13.36],
                    zoom_start=7,
                    tiles='OpenStreetMap')
```

```python
locations = {
    "sebastian": (52.5164293, 13.4780252),
    "oli": (52.4719915, 13.4306577),
    "lars": (51.331223, 9.425090),
    "andrey": (52.4629993, 13.48964776),
    "jorge": (52.6196931, 13.486191)
}
```

```python
for person, coordinates in locations.items():
    marker = folium.Marker(coordinates, popup=person.capitalize(),
            icon=folium.Icon(icon='star',
            color='orange'))
    marker.add_to(germany)
```
